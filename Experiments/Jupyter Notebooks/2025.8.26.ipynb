```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 1/(x*x*x*np.tan(np.sqrt(11)*np.pi*x))

x=np.arange(1,1000001,1)
y=f(x)


fig=plt.figure(figsize=(20,10),dpi=150)
ax=fig.add_subplot()

ax.scatter(x, y, s=1, color='black')

ax.set_xlim(0,1000000)
ax.set_ylim(0,0.00000000001)

plt.grid(True)
plt.show()

```
